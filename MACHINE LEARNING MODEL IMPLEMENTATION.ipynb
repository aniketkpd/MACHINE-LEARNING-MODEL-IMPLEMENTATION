# Car Buyer Predictor Using Machine Learning
# Tells wheather a person will buy a car or not

```python
import pandas as pd

# Loading the dataset
data = pd.read_csv('/content/Social_Network_Ads.csv')
print(data)
```

```python
# We Will find that either the person will by a new car or not.

# Suppose a person created a car -- marketed it on social media and the data is with us.

# 0 - not purchased
# 1 - purchased
```

```python
# Here
# Features/Input - 'User ID', 'Gender', 'Age', 'EstimatedSalary'
# Labels/Target Column/Output - 'Purchased'
```

```python
# Execution steps
# 1.Data-Preprocessing
# 2.Logistic regression
# 3.Evauluate
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dividing data into input(x) and output(y)

# data.iloc[ : , [2,3] ] #selecting only required columns
data.iloc[ : , [2,3] ].values #Convert the tables values into an array format

# implementing this concept to divide the datasetinto x and y
```

```python
x = data.iloc[ : ,2:4]
y = data.iloc[ : ,4]
```

```python
# Dividing data into training and testing
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state= 0)
```

```python
#Checking how data is divided as intended
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#Scalling between a particular range
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Creating object of standardscalar to scale
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
```

```python
x_test = sc.fit_transform(x_test)
```

```python
x_test
```

```python
x_test
```

```python
# Applying Logistic Regresttion
from sklearn.linear_model import LogisticRegression

#when we create the object of this LogisticRegression class, that object will be called our model

#Creating our model/Object

log_reg = LogisticRegression(random_state= 0 )

#Training the model
log_reg.fit(x_train,y_train)
```

```python
# Model has learnt now, now it can predict

y_pred = log_reg.predict(x_test)
# predicted data
print(y_pred)
```

```python
# Actual data
print(y_test)
```

```python
# Scatter Plot for Actual Data
plt.scatter(x_test[ : ,0],y_test)
```

```python
# Scatter Plot for Predicted Data
plt.scatter(x_test[ : ,0],y_pred)
```

```python
# Combined Scatter Plot
plt.scatter(x_test[ : ,0],y_test, c=y_pred)
```

```python
# Checking accuracy of the model.

from sklearn.metrics  import accuracy_score, confusion_matrix

print("Accuracy : ",accuracy_score(y_test,y_pred))
```

```python
# Printing confusion matrix

cf = confusion_matrix(y_test,y_pred)
print(cf)
```
